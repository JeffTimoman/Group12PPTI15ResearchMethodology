```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('305medium.jpeg')

# Convert the image to YUV color space
image_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

# Split the image into Y, U, and V components
Y, U, V = cv2.split(image_yuv)

# Plotting the histograms of Y, U, and V components
plt.figure(figsize=(15, 5))

# Plot Y component
plt.subplot(1, 3, 1)
plt.hist(Y.ravel(), bins=256, color='black', alpha=0.5)
plt.title('Y Component Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency')

# Plot U component
plt.subplot(1, 3, 2)
plt.hist(U.ravel(), bins=256, color='blue', alpha=0.5)
plt.title('U Component Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency')

# Plot V component
plt.subplot(1, 3, 3)
plt.hist(V.ravel(), bins=256, color='red', alpha=0.5)
plt.title('V Component Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('305medium_rembg.jpeg')

# Convert the image from BGR to YUV
yuv_image = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

# Extract the Y, U, and V channels
Y, U, V = cv2.split(yuv_image)

# Create empty arrays for U and V channels in RGB space
U_rgb = np.zeros_like(image)
V_rgb = np.zeros_like(image)

# Assign U and V channels to corresponding RGB channels for visualization
U_rgb[..., 1] = cv2.normalize(U, None, 0, 255, cv2.NORM_MINMAX)  # Green channel
V_rgb[..., 2] = cv2.normalize(V, None, 0, 255, cv2.NORM_MINMAX)  # Red channel

# Plot the channels
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(cv2.cvtColor(Y, cv2.COLOR_GRAY2RGB))
axs[0].set_title('Y Channel')
axs[0].axis('off')

axs[1].imshow(U_rgb)
axs[1].set_title('U Channel (mapped to Green)')
axs[1].axis('off')

axs[2].imshow(V_rgb)
axs[2].set_title('V Channel (mapped to Red)')
axs[2].axis('off')

plt.show()
```
